```python
#ensure annotations
def product(a:int, b:int) -> int:
    return a*b
```

```python
product(2,3)
product(2,"3")
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def product(a:int, b:int) -> int:
    return a*b
```

```python
product(2,3)
product(2,"3")
```
